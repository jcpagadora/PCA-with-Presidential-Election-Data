## Principal Components Analysis

In this notebook, we go over principal components analysis as a dimensionality reduction tool, which is especially useful if our data comes in with many features, and we believe that the data can be sufficiently summarized with a lower-rank data matrix.

The simplest way to derive the PCA is through the __singular value decomposition (SVD)__ and the Eckart-Young Theorem.

Suppose our data is rectangular and quantitative, and let $\mathbb{X}\in\mathbb{R}^{n\times d}$ denote our data matrix, where $n,$ the number of rows, is the number of samples, and $d,$ the number of columns, is the number of features in our data.

To start, let us first import all our necessary libraries and generate a two-dimensional dataset. Note that PCA is sensitive to location and scale, so before doing PCA on any dataset, make sure to normalize it first.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
```

```python
X = np.random.multivariate_normal(mean=[0,0], cov=[[3, 2], [2, 3]], size=150)

# Visualize the dataset
sns.set()
sns.scatterplot(X[:,0], X[:,1]);
plt.axvline(0);
plt.axhline(0);
```

First, we normalize the dataset $\mathbb{X}.$ Now let us take the SVD: $\mathbb{X}=U\Sigma V^T,$ where $\Sigma$ is diagonal, and $U$ and $V$ are orthogonal. Recall the intuition that if $\mathbb{X}$ were a transformation, then its action rotates Euclidean coordinates through $U,$ then $\Sigma$ scales, then $V^T$ rotates back.

Simply put, principal components analysis (PCA) just rotates the coordinate axes so that axis 1 explains the most variance of the data, axis 2 explains the 2nd most variance, and so on. In fact, the total variance of $\mathbb{X}$ can be decomposed:
$$\mathrm{Var}(\mathbb{X})=\frac{1}{n}\sum_{i=1}^d \sigma_i^2,$$
where $\sigma_i$ denotes the $i$-th singular value of $\mathbb{X}.$

The __principal components__ of $\mathbb{X}$ are the axes corresponding to the singular values. For example, the first principal component of $\mathbb{X}$ is the axis corresponding to $\sigma_1.$ Therefore from the decomposition above, we can see exactly how much of the variance each principal component explains.

Concretely, the principal component matrix is defined by $P:=U\Sigma,$ and the $i$-th principal component is just the $i$-th row of $P.$

```python
# The total variance of X is the sum of the column variances
for i in range(X.shape[1]):
    X[:,i] = (X[:,i] - np.mean(X[:,i])) / np.std(X[:,i])
n, d = X.shape
total_var_X_direct = np.sum([np.var(X[:,i]) for i in range(d)])
total_var_X_direct
```

```python
# Using the decomposition using singular values above
u, s, vt = np.linalg.svd(X)
total_var_X_sing_vals = np.sum(s ** 2) / n
total_var_X_sing_vals
```

```python
# We can display how much of the variance is explained by each principal component.
var_explained_by_pc1 = s[0] ** 2 / n
var_explained_by_pc2 = s[1] ** 2 / n
print("Variance explained by 1st PC: ", var_explained_by_pc1)
print("Variance explained by 2nd PC: ", var_explained_by_pc2)
```

```python
# Get the principal component matrix
sns.set()
plt.figure(figsize=(6, 6))
sns.scatterplot(X[:,0], X[:,1]);
plt.axvline(0);
plt.axhline(0);

# We plot the axes directions to visualize the actual principal components on the dataset. These are precisely the rows of the
# pincipal component matrix. Note that for more accurate scaling of these principal components, we scale by the corresponding
# variance explained by that principal component.
plt.arrow(0, 0, s[0]**2/n * vt[0,0], s[0]**2/n * vt[0,1], color='black', width=0.01, head_width=0.1, head_length=0.1);
plt.arrow(0, 0, s[1]**2/n * vt[1,0], s[1]**2/n * vt[1,1], color='black', width=0.01, head_width=0.1, head_length=0.1);
```

The Eckart-Young Theorem states that the closest lower-rank matrix to $\mathbb{X}$ in the Frobenius-norm is the one obtained by throwing away the smallest singular values from $\Sigma.$ For example, if $\mathbb{X}$ were rank $d,$ then to get the best rank $d-k$ approximation to $\mathbb{X},$ we simply throw away the smallest $k$ singular values of $\mathbb{X}.$

Thus, we can also perform dimensionality reduction on a data set using principal components analysis as well, by simply looking at those most important principal components of a data set.

Now, let's apply what we know about PCA to a real high-dimensional dataset to gain some better insights that would have been impossible without performing PCA. Let's look at presidential election data, which we obtain from the table on the following Wikipedia page: https://en.wikipedia.org/wiki/List_of_United_States_presidential_election_results_by_state

We will use URL requests and Beautiful Soup to extract the data from this page.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_United_States_presidential_election_results_by_state'
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

# Find all tables on the page, and observe that the table we want is the first table on the page
table = soup.find_all("tbody")[0]
```

```python
rowSoup = table.findChildren("tr")
```

```python
# First, we will gather the years to be the "features" of our dataset, as in the table on the page
yearSoup = rowSoup[0].findChildren("th")[1:]
years = []
for y in yearSoup[:-1]:
    years.append(y.findAll(text=True)[0])
    
# Initialize the dataframe "elections" with years as the features
elections = pd.DataFrame(years)
```

```python
# Then, gather the names of all the states in the order given in the table, namely, alphabetical
# After looking through the HTML, we see that the actual state names are in the 1st through 52nd indices of rowSoup,
# where one row repeates the word 'State\n', so we ignore it
states = []
for row in rowSoup[1:53]:
    state = row.findAll(text=True)[1]
    if state != 'State\n':
        states.append(state)
states
```

```python
# Initialize the pandas data frame as "elections"
elections = pd.DataFrame(index=states, columns=years)
elections.head(5)
```

```python
# Fill out the dataframe with the corresponding data. Since we are using rowSoup, we still have to deal with the extraneous
# row labeled 'State', which lists all the years again. Observe that in this case, the obtained row is [], an empty list.
i0, i1 = 0, 0
while i0 < 51:
    row = rowSoup[i1+1].findChildren("td")
    if row == []:
        i1 += 1
        continue
    for j in range(len(years)):
        elections.iloc[i0,j] = row[j].findAll(text=True)[0]
    i0, i1 = i0 + 1, i1 + 1
```

```python
elections.head(10)
```

For simplicity, let us look at the years from 1972 to 2016. Then, we are able to have every state participate in every election, voting either Republican (R) or Democrat (D). We then quantify the data, in particular, we set replace every R with 0 and every D with 1.

```python
elections_since_1972 = elections.loc[:,'1972':'2016']
elections_since_1972 = elections_since_1972.replace({'R':0, 'D':1, 'R\n':0, 'D\n':1})
elections_since_1972.head(10)
```

```python
# To prepare for our PCA, we will center and scale the data
elections_since_1972_norm = elections_since_1972.apply(lambda x: (x - np.mean(x)) / np.std(x), axis=0)
```

```python
# Take the SVD of the normalized dataset, then take the first 2 principal components
u, s, vt = np.linalg.svd(elections_since_1972_norm)
pcs = elections_since_1972_norm @ vt.T
first_2_pcs = pcs[[0,1]]
```

```python
# We plot the first 2 principal components
plt.figure(figsize=(6, 4))
sns.scatterplot(data = first_2_pcs, x = 0, y = 1)
plt.xlabel("pc1")
plt.ylabel("pc2");
```

We should have 51 data points on our plot, but that is not noticeable from the graph above. Indeed, there is overfitting in our plot, where some points lie on top of other points, so we add jitter to help our visualization.

```python
# We plot the first 2 principal components
first_2_pcs_jittered = first_2_pcs + np.random.normal(0, 0.1, size = (len(first_2_pcs), 2))
plt.figure(figsize=(6, 4))
sns.scatterplot(data = first_2_pcs_jittered, x = 0, y = 1)
plt.xlabel("pc1")
plt.ylabel("pc2");
```
